```python
import warnings
warnings.filterwarnings('ignore')
categorical_cols = df.select_dtypes(include='object')
categorical_cols
```

```python
##Day
sns.countplot(data=categorical_cols, x='Day', order=df['Day'].value_counts().index)
plt.title("Count of Sales by Day")
plt.xlabel("Day of Week")
plt.ylabel("Count of Sales")
plt.show()
```

```python
sns.countplot(data=categorical_cols, x='City', color='violet', order=df['City'].value_counts().index)
plt.title("Count of Sales by City")
plt.xlabel("City")
plt.ylabel("no of sale")
plt.show()
```

```python
sns.countplot(data=categorical_cols, x='BU', color='green', order=df['BU'].value_counts().index)
plt.title("Count of Sales by Business Unit")
plt.xlabel("Business Unit")
plt.ylabel("Count of Sales")
plt.show()
```

```python
sns.countplot(data=categorical_cols, x='Brand', palette='crest', order=df['Brand'].value_counts().index)
plt.title("Count of Sales by Brand")
plt.xlabel("Brand")
plt.ylabel("Count of Sales")
plt.show()
```
