###### 2. Select only the columns relevant for billing: ['PatientID', 'Department', 'Doctor', 'BillAmount'].

```python
Billing=Patient_data[['PatientID', 'Department', 'Doctor', 'BillAmount']]
Billing
```

###### 3. Drop administrative columns like ['ReceptionistID', 'CheckInTime'].

```python
Patient_data.drop(columns=['ReceptionistID', 'CheckInTime'],inplace=True)
Patient_data
```

###### 4. Use groupby to find total bill amount per department.

```python
Patient_data.groupby('Department')['BillAmount'].sum().reset_index()
```

###### 5. Remove duplicate patient records based on PatientID.

```python
Patient_data.drop_duplicates(subset=['PatientID'],ignore_index=True,inplace=True)
Patient_data
```

###### 6. Fill missing BillAmount values with the mean bill amount.

```python
Patient_data.BillAmount.mean()
```

```python
Patient_data['BillAmount'].fillna(Patient_data['BillAmount'].mean(),inplace=True)
Patient_data
```
